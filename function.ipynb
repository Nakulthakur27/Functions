3. Explain why typecasting errors can break ML pipelines

Typecasting is the process of converting data from one type to another (e.g., string to integer or float).

Why typecasting errors are problematic in Machine Learning:

1.Mathematical operations fail


 .ML algorithms require numeric data.                                     
 .If numbers are stored as strings, calculations cannot be performed correctly.

```python
"5" * "10"   # Error
```

2.Incorrect results

.String operations behave differently from numeric operations.

```python
"5" + "10"   # Output: "510"
5 + 10       # Output: 15
```

3.Model training errors

.Most ML libraries (such as NumPy, Pandas, and Scikit-learn) expect numerical features.

.String values may cause the model to fail during training.

4.Data quality issues

.Incorrect data types can lead to wrong predictions and unreliable analysis.

5.Pipeline failures

.A single type mismatch can stop an entire ETL (Extract, Transform, Load) or ML workflow.

Example:


```python
price = "499.99"
quantity = "5"

# Without typecasting
total = price * quantity   # Error
```

After proper typecasting:

```python
price = "499.99"
quantity = "5"
total = float(price) * int(quantity)
print(total)
```

Conclusion:

Typecasting ensures data is in the correct format for calculations, analysis, and machine learning models. Incorrect data types can cause errors, inaccurate results, and complete pipeline failures.

Question 5:

 Given: sales = [12000, 15000, 11000, 18000, 17000] Using a loop:
  1. Calculate total revenue.
  2. Count how many months had sales above 15000.
  3. Find maximum sales without using built-in max()

Answer5:

```python
sales = [12000, 15000, 11000, 18000, 17000]

total_revenue = 0
months_above_15000 = 0
max_sales = sales[0]  # Start with first element as initial maximum

for sale in sales:
    # 1. Add each sale to total
    total_revenue += sale

    # 2. Check if sale is strictly above 15000
    if sale > 15000:
        months_above_15000 += 1

    # 3. Update max if current sale is greater
    if sale > max_sales:
        max_sales = sale

print(f"Total Revenue: ₹{total_revenue}")
print(f"Months above ₹15,000: {months_above_15000}")
print(f"Maximum Sales: ₹{max_sales}")
```

Question 9:

Write a generator that yields squares of numbers from 1 to 1,000,000.

Explain difference between:

● List comprehension

● Generator expression
Why are generators preferred in big data scenarios?

Answer9:


Generator that yields squares from 1 to 1,000,000

```python
def square_generator():
    for i in range(1, 1_000_001):
        yield i * i

# Example usage
gen = square_generator()

print(next(gen))  # 1
print(next(gen))  # 4
print(next(gen))  # 9
```

Difference Between List Comprehension and Generator Expression

1. List Comprehension


    Creates and stores the entire list in memory immediately.

```python
squares_list = [x**2 for x in range(1, 1_000_001)]
```

Characteristics:

.Faster access to elements.

.Uses more memory.

.Suitable for small to medium datasets.

.All values are generated at once.

Example:

```python
print(squares_list[0])  # 1
print(squares_list[1])  # 4
```

2. Generator Expression

Generates values lazily (on demand).

```python
squares_gen = (x**2 for x in range(1, 1_000_001))
```

Characteristics:

.Memory efficient.

.Values are generated only when needed

.Suitable for large datasets or streams.

.Can be iterated only once.

Example:

```python
print(next(squares_gen))  # 1
print(next(squares_gen))  # 4
```

Why Generators Are Preferred in Big Data Scenarios

Suppose you need squares of 1,000,000 numbers:

List Comprehension

```python
squares = [x**2 for x in range(1, 1_000_001)]
```

.Creates 1,000,000 values immediately.

.Consumes significant memory.

.Can cause memory issues with much larger datasets.

Generator

```python
squares = (x**2 for x in range(1, 1_000_001))
```

.Stores only the current value and state.

.Memory usage remains almost constant.

.Can process datasets larger than available RAM.


Real-World Big Data Uses

Generators are commonly used for:

    .Reading huge log files line by line.

    .Processing millions of database records.

    .Streaming sensor/IoT data.

    .ETL (Extract–Transform–Load) pipelines.

    .Machine learning data loaders.

    .Processing large CSV or JSON files.

Example:

```python
def read_large_file(filename):
    with open(filename) as f:
        for line in f:
            yield line
```

Instead of loading a 10 GB file into memory, the generator processes one line at a time, making it highly scalable and memory efficient.

Question 10:

 Design a class:
class Product:
With:

● Attributes: name, price, quantity

● Method: calculate_revenue()

Create 3 objects and print their revenue.

Answer10:

```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def calculate_revenue(self):
        return self.price * self.quantity


# Creating 3 objects
product1 = Product("Laptop", 50000, 5)
product2 = Product("Mobile", 20000, 10)
product3 = Product("Headphones", 1500, 20)

# Printing revenue
print(product1.name, "Revenue =", product1.calculate_revenue())
print(product2.name, "Revenue =", product2.calculate_revenue())
print(product3.name, "Revenue =", product3.calculate_revenue())
```

The calculate_revenue() method calculates revenue using:

Revenue=Price×Quantity
